# Breakout Strategy Assignment

This notebook runs a long-only breakout strategy with a rolling walk-forward framework, exports the blotter and ledger, and prepares website-ready artifacts for GitHub Pages.


## Strategy Summary

- Breakout rule: buy when the closing price breaks above the highest high from the prior lookback window.
- Exit logic: ATR-based stop-loss, ATR-based profit target, or timeout.
- Walk-forward design: one year of training, followed by one quarter of out-of-sample testing.
- Data source policy: try `shinybroker` first; if no local IBKR/TWS session is available, fall back to a public Yahoo chart endpoint.
